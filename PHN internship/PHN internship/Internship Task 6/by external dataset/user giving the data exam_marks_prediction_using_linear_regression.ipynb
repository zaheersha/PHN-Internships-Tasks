```python
dataset = pd.read_csv('std_marks_data.csv') #load the data set 
```

```python
print(dataset.shape)
print(dataset.head(5))
print(dataset.tail(5))
```

```python
#finding and removing the nan values from our feature X

dataset.columns[dataset.isna().any()] 

 #for findint the index where the nan is present amoung the compleat dataset
```

```python
dataset.hours = dataset.hours.fillna(dataset.hours.mean())
dataset
```

```python
#segrate dataset into input x and output y
# input as expect last column remaining coloumns are outputs
X = dataset.iloc[:,:-1].values
print(X.shape)
X
```

```python
#segrate dataset into input x and output y
Y = dataset.iloc[:,-1].values
Y
```

```python
#training the data set using LR
model = LinearRegression()
model.fit(X,Y)
```

```python
#predict 

hours = int(input("enter hours  : "))
age = int(input("enter  age : "))
internet = int(input("enter internet is available or not (0/1)  : "))
a=[[hours,age,internet]]
#a=[[9,20,1]]
PredictemodelResult = model.predict(a)
print(PredictemodelResult)
```
